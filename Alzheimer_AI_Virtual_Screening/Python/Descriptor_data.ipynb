```python
import pandas as pd
import pubchempy as pcp
```

```python
df = pd.read_csv("/content/drive/MyDrive/Alzheimer_AI_Virtual_Screening/Results/Docking_Results.csv")

df.head()
```

Fetch Molecular Descriptors Automatically

```python
descriptor_data = []

for ligand in df["Ligand"]:
    try:
        compound = pcp.get_compounds(ligand, "name")[0]

        descriptor_data.append({
            "Ligand": ligand,
            "PubChem CID": compound.cid,
            "Molecular Weight": compound.molecular_weight,
            "XLogP": compound.xlogp,
            "TPSA": compound.tpsa,
            "H Bond Donor": compound.h_bond_donor_count,
            "H Bond Acceptor": compound.h_bond_acceptor_count,
            "Rotatable Bonds": compound.rotatable_bond_count
        })

        print(f"✓ {ligand}")

    except:
        print(f"✗ {ligand} not found")
```

Create the Descriptor Table

```python
descriptor_df = pd.DataFrame(descriptor_data)

descriptor_df.head()
```

```python
descriptor_df.to_excel(
    "Molecular_Descriptors.xlsx",
    index=False
)
```

Merge with Docking Results

```python
final_df = pd.merge(
    df,
    descriptor_df,
    on="Ligand",
    how="left"
)

final_df.head()
```

Extract the crystal ligand

```python
from Bio.PDB import PDBParser, PDBIO, Select

class LigandSelect(Select):
    def accept_residue(self, residue):
        return residue.get_resname() == "E20"   # Replace if ligand name differs

parser = PDBParser(QUIET=True)
structure = parser.get_structure("protein", "4EY7.pdb")

io = PDBIO()
io.set_structure(structure)
io.save("Donepezil_crystal.pdb", LigandSelect())
```

Calculate RMSD

```python
from Bio.PDB import PDBParser, Superimposer

parser = PDBParser(QUIET=True)

crystal = parser.get_structure("crystal", "/content/Donepezil_crystal.pdb")
docked = parser.get_structure("docked", "/content/Donepezil_redocked.pdb")

crystal_atoms = [atom for atom in crystal.get_atoms() if atom.element != "H"]
docked_atoms = [atom for atom in docked.get_atoms() if atom.element != "H"]

sup = Superimposer()
sup.set_atoms(crystal_atoms, docked_atoms)

print("Validation RMSD:", round(sup.rms,3), "Å")
```
